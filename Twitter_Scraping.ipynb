```python
# Import Libraries
import pandas as pd
import numpy as np
import re
```

```python
keywords = [
    "RUU KUHP",
    "KUHP",
    "RKUHP",
    "revisi KUHP",
    "revvisi RUU",
    "Hukum Pidana",
    "Undang-Undang Hukum Pidana",
    "revisi RKUHP",
]
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list1 = []

# Define the time range you want to loop through
start_year = 2015
end_year = 2023

for keyword in keywords:
    for year in range(start_year, end_year + 1):
        # Construct the since and until dates for the current iteration
        since_date = f"{year}-01-01"
        until_date = f"{year+1}-01-01"

        # Scrape tweets within the specified time range
        for i, tweet in enumerate(sntwitter.TwitterSearchScraper(keyword + f' since:{since_date} until:{until_date}').get_items()):
            if i > 200:
                break
            tweets_list1.append([tweet.date, tweet.rawContent, keyword])

# Creating a dataframe from the tweets list
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Text', 'Keyword'])
```
